```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL   
url = "https://knu.ua/news"

#    
response = requests.get(url)
response.encoding = 'utf-8'  #       

#   
if response.status_code == 200:
    #  ' BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #   ,    
    headlines = []
    for item in soup.find_all('a', class_='b-menu__link'):
        title = item.get_text(strip=True)
        headlines.append({'': title})
    
    #  DataFrame   
    df = pd.DataFrame(headlines)
    
    #   CSV-
    df.to_csv('knu_headlines.csv', index=False, encoding='utf-8')
    print("    knu_headlines.csv")
else:
    print(f"     .  : {response.status_code}")
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Generate sample data for Excel
periods = 2

data = {
    "Date": pd.date_range(start="2024-09-01", periods=periods, freq="D"),
    "Value": [round(1000 * (1 + 0.01 * i)) for i in range(periods)]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Save to an Excel file
file_path = 'sample_data2.xlsx'
df.to_excel(file_path, index=False)

file_path
```

```python
import xml.etree.ElementTree as ET

#   
units = [
    {"name": "Alpha", "quantity": 100, "status": "Active"},
    {"name": "Bravo", "quantity": 80, "status": "Inactive"},
    {"name": "Charlie", "quantity": 120, "status": "Active"},
]

#   
root = ET.Element("Units")

#    XML
for unit in units:
    unit_element = ET.SubElement(root, "Unit")
    
    name_element = ET.SubElement(unit_element, "Name")
    name_element.text = unit["name"]
    
    quantity_element = ET.SubElement(unit_element, "Quantity")
    quantity_element.text = str(unit["quantity"])
    
    status_element = ET.SubElement(unit_element, "Status")
    status_element.text = unit["status"]

#      
tree = ET.ElementTree(root)
tree.write("units.xml", encoding="utf-8", xml_declaration=True)

print("XML- : units.xml")
```
